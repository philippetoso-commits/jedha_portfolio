```python
import boto3
import os
from dotenv import load_dotenv

# 1. Configuration et chargement des identifiants
load_dotenv()
AWS_ACCESS_KEY = os.getenv("AWS_ACCESS_KEY_ID") or os.getenv("AWS_ACCESS_KEY")
AWS_SECRET_KEY = os.getenv("AWS_SECRET_ACCESS_KEY") or os.getenv("AWS_SECRET_KEY")
BUCKET_NAME = os.getenv("BUCKET_NAME")
FILENAME = "booking_final_gps.csv"

# 2. Connexion et Upload vers S3
print(f" Envoi de '{FILENAME}' vers le bucket S3 '{BUCKET_NAME}'...")

session = boto3.Session(
    aws_access_key_id=AWS_ACCESS_KEY,
    aws_secret_access_key=AWS_SECRET_KEY,
    region_name="eu-west-3" # Vérifiez que c'est bien votre région
)
s3 = session.client('s3')

try:
    s3.upload_file(FILENAME, BUCKET_NAME, FILENAME)
    print(f" Succès ! Le fichier est maintenant disponible sur S3 : s3://{BUCKET_NAME}/{FILENAME}")
except Exception as e:
    print(f" Erreur lors de l'upload : {e}")
```
